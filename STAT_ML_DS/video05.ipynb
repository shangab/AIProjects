```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('./data/student_spending.csv')
df.head(3)
```

```python
df["foodz"] = (df["food"] - df["food"].mean()) / df["food"].std()

df.head(10)
```

```python
df2 = df[["food", "foodz"]]

df2.head(10)
```

```python
outliers = df2[(df2["foodz"] > 3) | (df2["foodz"] < -3)]
outliers
```

```python
no_outliers = df2[(df2["foodz"] <= 3) & (df2["foodz"] >= -3)]
```
