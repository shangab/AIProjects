```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./data/DiamondsPrices_cleaned.csv')
data.head()
```

```python
data.shape
```

```python
clarity = pd.get_dummies(data['clarity'], drop_first=False, dtype=int)
clarity.head(10)
```

```python
data = pd.concat([clarity, data], axis=1)
data.head()
```

```python
data = data.drop(['clarity'], axis=1)
data.head()
```

```python
X = data.drop(['price'], axis=1)
y = data['price']
```

```python
X.head(3)
```

```python
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42)
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
score = model_lr.score(X_test, y_test)
score
```

```python
model_lso = Lasso()
model_lso.fit(X_train, y_train)
score = model_lso.score(X_test, y_test)
score
```

```python
model_dtr = DecisionTreeRegressor()
model_dtr.fit(X_train, y_train)
score = model_dtr.score(X_test, y_test)
score
```

```python
from xgboost import XGBRegressor
model_xgb = XGBRegressor()
model_xgb.fit(X_train, y_train)
score = model_xgb.score(X_test, y_test)
score
```

```python
# input = ['I1', 'IF', 'SI1', 'SI2', 'VS1', 'VS2', 'VVS1', 'VVS2', 'carat',
#    'depth', 'table', 'x', 'y', 'z']
# carat : 0.2 - 3
# depth : 55 - 70
# table : 52 - 68
# x : 3 - 9
# y : 3 - 9
# z : 2 - 5
# Output: Price
def predict_price(model, clarity, carat, depth, table, x, y, z):
    columns = ['I1', 'IF', 'SI1', 'SI2', 'VS1', 'VS2', 'VVS1',
               'VVS2', 'carat', 'depth', 'table', 'x', 'y', 'z']
    x_input = [0, 0, 0, 0, 0, 0, 0, 0, carat, depth, table, x, y, z]

    clarity_index = columns.index(clarity)
    x_input[clarity_index] = 1
    x_input = np.array(x_input).reshape(1, -1)

    price = model.predict(x_input)
    return price[0]
```
